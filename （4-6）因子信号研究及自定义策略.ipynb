# 因子信号研究及自定义策略（4-6）

```python
#读取数据：分析用数据BTCUSDT:binance，周期1H，起始时间20180601000000
import warnings
warnings.filterwarnings('ignore')
from jaqs_fxdayu.data.dataservice import LocalDataService
ds = LocalDataService()
from time import time
## 加freq参数

path = r'./min_data/VnTrader_1Min_Db'
props = {'fields': 'open,high,low,close,volume','symbol': \
         'BTCUSDT:binance',
         'freq': '30Min,1H',
         'start_date':20180601000000}

Time_dict = ds.bar_reader(path,props) #读取数据
from jaqs_fxdayu.data.hf_dataview import HFDataView

dv1H = HFDataView()

dv1H.create_init_dv(Time_dict['1H'].dropna().set_index(["trade_date","symbol"]))
```

```python
from jaqs_fxdayu.research import TimingDigger
#定义分析函数
def TimingSignal(td, dv, long='long', short='short',closeLong='closeLong', closeShort='closeShort'\
                 , mhp=None,sl=None,sp=None):
    # step 1：实例化TimingDigger 通过output_folder和output_format指定测试报告的输出路径和输出格式，通过signal_name指定绩效文件名称

    #多空信号分别计算一遍 输出汇总结果
    td.process_signal(
        enter_signal=dv.get_ts(long),
        exit_signal=dv.get_ts(closeLong),
        sig_type="long",  # 信号类型 long/short
        price=dv.get_ts('close'),
        max_holding_period=mhp,  # 最大持有天数 可为空
        stoploss=-sl,  # 止损百分比 负数 可为空
        stopprofit=sp,  # 止盈百分比 正数 可为空
    )

    td.process_signal(
        enter_signal=dv.get_ts(short),
        exit_signal=dv.get_ts(closeShort),
        sig_type="short",  # 信号类型 long/short
        price=dv.get_ts("close"),
        max_holding_period=mhp,  # 最大持有天数 可为空
        stoploss=-sl,  # 止损百分比 负数 可为空
        stopprofit=sp,  # 止盈百分比 正数 可为空
    )
```

```python
import matplotlib.pyplot as plt
import pandas as pd
#画图观察，思路：先从最基础的均线策略入手，然后想办法优化均线策略
fig,ax = plt.subplots(figsize=(16,4))
ax.plot(dv1H.get_ts("close",date_type="datetime"))
ax.plot(dv1H.get_ts("close",date_type="datetime").rolling(5).mean(),color='r')
ax.plot(dv1H.get_ts("close",date_type="datetime").rolling(20).mean(),color='g')
ax.plot(dv1H.get_ts("close",date_type="datetime").rolling(72).mean(),color='b')
plt.show()
```

```python
#先查看基础均线策略效果
# 进场信号
long = dv1H.add_formula("long","If((Ts_Mean(close,5)>Ts_Mean(close,20)) && (Delay(Ts_Mean(close,5),1)<=Delay(Ts_Mean(close,20),1)),2,0)", add_data=True)
short = dv1H.add_formula("short","If((Ts_Mean(close,5)<Ts_Mean(close,20)) && (Delay(Ts_Mean(close,5),1)>=Delay(Ts_Mean(close,20),1)),-2,0)", add_data=True)

# 出场信号
close_long = dv1H.add_formula("closeLong","If(short==-2,1,0)", add_data=True)
close_short = dv1H.add_formula("closeShort","If(long==2,-1,0)", add_data=True)

tdMa = TimingDigger(output_folder=".", output_format='pdf',signal_name='Ma')
TimingSignal(tdMa, dv1H, mhp=100, sl=0.05, sp=0.3)
tdMa.create_event_report(sig_type="long_short")
```

```python
#可以看出基础均线效果不错
#优化均线策略思路1：用长期均线的斜率来过滤掉长期下跌趋势中的反弹
#定义长期均线的斜率，这里用72作为长期周期，因为一天24小时，用三天的平均作为长期均线
longma72del = dv1H.add_formula("longma72del","Ts_Mean(close,72)-Delay(Ts_Mean(close,72),1)", add_data=True)
#画图查看并查看统计指标
dv1H.get_ts("longma72del",date_type="datetime").plot()
dv1H.get_ts("longma72del").hist()
plt.show()
dv1H.get_ts("longma72del").describe()
```

```python
#由上图可以看出，大部分数值处于-5到5之间，那么我们可以用来过滤，小于-5为强下跌趋势，不做多，大于5为强烈上涨趋势，不做空
longmaf = dv1H.add_formula("longmaf","If((long==2)&&(longma72del>-5),2,0)", add_data=True)
shortmaf = dv1H.add_formula("shortmaf","If((short==-2)&&(longma72del<5),-2,0)",add_data=True)

tdMaf = TimingDigger(output_folder=".", output_format='pdf',signal_name='Maf')
TimingSignal(tdMaf, dv1H,'longmaf','shortmaf','closeLong','closeShort', 100,0.05,)
tdMaf.create_event_report(sig_type="long_short")
```

```python
#可以看出效果略有提升，看看进出场信号是如何
import mpl_finance as mpf

def draw_candle_signal(candle, longX, longY,shortX,shortY):
    fig, (ax, ax1) = plt.subplots(2,1,sharex=True, figsize=(18,10))
    ax.grid(True)
    mpf.candlestick_ohlc(ax, candle, width=0.3, colorup='r', colordown='g',alpha=1.0)
    ax.scatter(longX, longY, color='b', s=100, marker='^')
    ax.scatter(shortX,shortY,color='magenta',s=100,marker='v')
    ax.xaxis_date ()
    ax1.bar(data1H.index, data1H.volume)
    plt.show()

from matplotlib.pylab import date2num

data1H = Time_dict['1H'].set_index('datetime')
data1H['candleTime'] = list(map(date2num, data1H.index))
candle = data1H.reindex_axis(["candleTime", "open", "high", "low", "close"], 1).values

longsignal = dv1H.get_ts('longmaf', date_type='datetime')
shortsignal = dv1H.get_ts('shortmaf', date_type='datetime')

longX = longsignal[longsignal['BTCUSDT:binance']==2].index
longY = data1H.loc[longX].low-100
shortX = shortsignal[shortsignal['BTCUSDT:binance']==-2].index
shortY = data1H.loc[shortX].high+100

draw_candle_signal(candle,longX,longY,shortX,shortY)
```

```python
#优化均线策略思路2：用RSI作为过滤条件
RSI6 = dv1H.add_formula("RSI6","Ta('RSI',0, open, high, low, close, volume,6)",overwrite=True,add_data=True)
RSI12 = dv1H.add_formula("RSI12","Ta('RSI',0, open, high, low, close, volume,12)",overwrite=True,add_data=True)
RSI24 = dv1H.add_formula("RSI24","Ta('RSI',0, open, high, low, close, volume,24)",overwrite=True,add_data=True)
RSI36 = dv1H.add_formula("RSI36","Ta('RSI',0, open, high, low, close, volume,36)",overwrite=True,add_data=True)
fig,ax = plt.subplots(figsize=(16,4))
ax.plot(dv1H.get_ts("RSI6",date_type="datetime").iloc[:500],color='r')
ax.plot(dv1H.get_ts("RSI12",date_type="datetime").iloc[:500],color='g')
ax.plot(dv1H.get_ts("RSI24",date_type="datetime").iloc[:500],color='b')
plt.show()
```

```python
#结合均线策略，只在RSI6>RSI24的时候做多，也只在相反的时候做空
longfilterma = dv1H.add_formula("longfilterma","If((long==2)&&(RSI12>RSI24),2,0)", add_data=True)
shortfilterma = dv1H.add_formula("shortfilterma","If((short==-2)&&(RSI12<RSI24),-2,0)",add_data=True)

tdMaf = TimingDigger(output_folder=".", output_format='pdf',signal_name='Maf')
TimingSignal(tdMaf, dv1H,'longfilterma','shortfilterma','closeLong','closeShort', 100,0.05,)
tdMaf.create_event_report(sig_type="long_short")
```

```python
#可以看出效果并不太好。思考是因为RSI比MA敏感，用敏感指标去过滤滞后指标效果并不好
#优化均线策略思路3：用成交量作为过滤条件，只有在成交量放大时候的信号才采纳，因为震荡会产生很多假信号，而震荡时往往成交量不会突然放大
#因为一天以24小时为周期，所以成交量应该在24小时内平均，不然会被交易量在特定时间的波动误导（比如特定时区的开收盘）
volumedelta = dv1H.add_formula("volumedel","If((Ts_Mean(volume,24)>Ts_Mean(volume,48)),1,0)", add_data=True)

longmav = dv1H.add_formula("longmav","If((long==2)&&(volumedel==1),2,0)", add_data=True)
shortmav = dv1H.add_formula("shortmav","If((short==-2)&&(volumedel==1),-2,0)",add_data=True)

tdMaV = TimingDigger(output_folder=".", output_format='pdf',signal_name='MaV')
TimingSignal(tdMaV, dv1H,'longmav','shortmav','closeLong','closeShort', 100,0.05,)
tdMaV.create_event_report(sig_type="long_short")
```

```python
#可以看出效果有提升，画出思路3的信号：
longsignal = dv1H.get_ts('longmav', date_type='datetime')
shortsignal = dv1H.get_ts('shortmav', date_type='datetime')

longX = longsignal[longsignal['BTCUSDT:binance']==2].index
longY = data1H.loc[longX].low-100
shortX = shortsignal[shortsignal['BTCUSDT:binance']==-2].index
shortY = data1H.loc[shortX].high+100

draw_candle_signal(candle,longX,longY,shortX,shortY)
```

```python
#以上是第一个因子：均线因子，的优化思路
#下面寻找第二个因子，并优化。从alpha101中的alpha002获取到灵感
alpha2_change = dv1H.add_formula('alpha2_C','-1*Correlation(Delta(Log(volume),2),((close-open)/open),6)',add_data=True)
alpha2_plot = dv1H.get_ts('alpha2_C',date_type='datetime')
close = dv1H.get_ts('close', date_type='datetime')

def plot_chart(close,alpha):
    fig,(ax,ax1) = plt.subplots(2,1,sharex=True, figsize=(15,8))
    ax.plot(close.iloc[:500])
    ax1.plot(alpha.iloc[:500])
    # plt.savefig('good')
    plt.show()
    
plot_chart(close,alpha2_plot)
```

```python
#上图可以看出原论文中的alpha002是非常差的因子，毫无作用。
#alpha002的思路是量价波动率的背离，那么我们如何更好的定义量价波动率的背离。也就是价格波动不大的时候，成交量突然放大。
#改进alpha002，成交量和价格的波动量要重新定义，用24小时均值的变化率的相关系数定义
alpha2_change = dv1H.add_formula('alpha2_C','-1*Correlation(Delta(Log(Ts_Mean(volume,24)),2),Delta(Log(Ts_Mean((high-low)/close,24)),2),6)',add_data=True)
alpha2_plot = dv1H.get_ts('alpha2_C',date_type='datetime')
close = dv1H.get_ts('close', date_type='datetime')

def plot_chart(close,alpha):
    fig,(ax,ax1) = plt.subplots(2,1,sharex=True, figsize=(15,8))
    ax.plot(close.iloc[:])
    ax1.plot(alpha.iloc[:])
    # plt.savefig('good')
    plt.show()
    
plot_chart(close,alpha2_plot)
```

```python
#毛刺太多，用均值平滑
alpha2mean = dv1H.add_formula("alpha2mean","Ts_Mean(alpha2_C,24)",add_data=True)
alpha2plotmean = dv1H.get_ts("alpha2mean",date_type="datetime")

plot_chart(close,alpha2plotmean)
```

```python
#由上图可以看出，这是相当好的预测因子，每次凸起之后都预示着一波趋势的产生。但是，我们不知道趋势产生的方向
#我认为，趋势产生的方向，是与之前的中期趋势相反，那么如何定义中期趋势
#定义中期趋势思路1：参考之前的计算，用中期均线的斜率作定义，用均值平滑波动
longma72delmean = dv1H.add_formula("longma72delmean","Ts_Mean(longma72del,5)", add_data=True)

longstdvolprice = dv1H.add_formula("longstdvolprice","If((longma72delmean<0)&&(alpha2mean>-0.6),2,0)",add_data=True)
shortstdvolprice = dv1H.add_formula("shortstdvolprice","If((longma72delmean>0)&&(alpha2mean>-0.6),-2,0)",add_data=True)

close_longstd = dv1H.add_formula("closeLongstd","If(longstdvolprice==-2,1,0)", add_data=True)
close_shortstd = dv1H.add_formula("closeShortstd","If(shortstdvolprice==2,-1,0)", add_data=True)

tdstdvolprice = TimingDigger(output_folder=".", output_format='pdf',signal_name='tdstdvolprice')
TimingSignal(tdstdvolprice, dv1H,'longstdvolprice','shortstdvolprice','closeLongstd','closeShortstd', 100,0.05,)
tdstdvolprice.create_event_report(sig_type="long_short")
```

```python
#效果还不错，胜率达到0.6，而且是相对高频的策略，画出进出场信号
longsignal = dv1H.get_ts('longstdvolprice', date_type='datetime')
shortsignal = dv1H.get_ts('shortstdvolprice', date_type='datetime')

longX = longsignal[longsignal['BTCUSDT:binance']==2].index
longY = data1H.loc[longX].low-100
shortX = shortsignal[shortsignal['BTCUSDT:binance']==-2].index
shortY = data1H.loc[shortX].high+100

draw_candle_signal(candle,longX,longY,shortX,shortY)
```

```python
#定义中期趋势思路2，用中期RSI的数值作区分，大于50为上涨趋势，小于50为下跌趋势
longstdvolrsi = dv1H.add_formula("longstdvolrsi","If((RSI36<50)&&(alpha2mean>-0.7),2,0)",add_data=True)
shortstdvolrsi = dv1H.add_formula("shortstdvolrsi","If((RSI36>50)&&(alpha2mean>-0.7),-2,0)",add_data=True)

close_longstd = dv1H.add_formula("closeLongstdrsi","If(longstdvolrsi==-2,1,0)", add_data=True)
close_shortstd = dv1H.add_formula("closeShortstdrsi","If(shortstdvolrsi==2,-1,0)", add_data=True)

tdstdvolrsi = TimingDigger(output_folder=".", output_format='pdf',signal_name='tdstdvolrsi')
TimingSignal(tdstdvolrsi, dv1H,'longstdvolrsi','shortstdvolrsi','closeLongstdrsi','closeShortstdrsi', 100,0.05,)
tdstdvolrsi.create_event_report(sig_type="long_short")
```

```python
#定义价格波动率的波动率
STD = dv1H.add_formula("STD","StdDev(close/Delay(close,1)-1, 20)",add_data=True)
dv1H.get_ts("STD",date_type="datetime").plot(figsize=(16,4))
dv1H.get_ts("STD",date_type="datetime").hist(bins=100,figsize=(16,4))
plt.show()
```

```python
#可以看出效果不太好，信号太多，胜率还在，但因为假信号太多导致盈利下降
#为了过滤掉假信号，加入价格波动率作为条件，因为我们要找出量变，价不变的时候，这时候意味着市场产生大的分歧

longstdvolrsi = dv1H.add_formula("longstdvolrsi","If((RSI36<50)&&(alpha2mean>-0.7)&&(STD<0.003),2,0)",add_data=True)
shortstdvolrsi = dv1H.add_formula("shortstdvolrsi","If((RSI36>50)&&(alpha2mean>-0.7)&&(STD<0.003),-2,0)",add_data=True)

close_longstd = dv1H.add_formula("closeLongstdrsi","If(longstdvolrsi==-2,1,0)", add_data=True)
close_shortstd = dv1H.add_formula("closeShortstdrsi","If(shortstdvolrsi==2,-1,0)", add_data=True)

tdstdvolrsi = TimingDigger(output_folder=".", output_format='pdf',signal_name='tdstdvolrsi')
TimingSignal(tdstdvolrsi, dv1H,'longstdvolrsi','shortstdvolrsi','closeLongstdrsi','closeShortstdrsi', 100,0.05,)
tdstdvolrsi.create_event_report(sig_type="long_short")
```

```python
#可以看出，加入波动率作为过滤条件后，胜率和盈亏比都有明显提升，而且交易频率也不低。画出进出场信号
longsignal = dv1H.get_ts('longstdvolrsi', date_type='datetime')
shortsignal = dv1H.get_ts('shortstdvolrsi', date_type='datetime')

longX = longsignal[longsignal['BTCUSDT:binance']==2].index
longY = data1H.loc[longX].low-100
shortX = shortsignal[shortsignal['BTCUSDT:binance']==-2].index
shortY = data1H.loc[shortX].high+100

draw_candle_signal(candle,longX,longY,shortX,shortY)
```

## 研究结论：

综上研究可以看出，效果比较好的是1：均线成交量综合策略，2：量价波动率背离策略

基本假设是成交量意味着分歧，一般成交量越大意味着分歧越大，也就意味着变盘的产生。

要提升策略的效果，需要结合不同的因子，过滤掉不想要的虚假信号。
